```python
import numpy as np
import pandas as pd
import scipy.stats as stat
import matplotlib as plt
from scipy.optimize import newton_krylov
from scipy.integrate import dblquad
from itertools import combinations, chain, repeat
from tqdm import trange
```

```python
# sort the data from smallest to largest
df = pd.read_csv('Book1.csv')
param_df = df['TDS'].to_frame().dropna(how='all').reset_index(drop=True)
param_df.sort_values(by='TDS', ignore_index=True, inplace=True)
param_df
```

```python
# find the empirical cdf
param_df['Probabilities'] = (param_df.index + 1)/(param_df.shape[0] + 1)
param_df
```

```python
# interpolate any value to find normal cdf?
np.interp([0.1, 0.3, 0.5], param_df['Probabilities'], param_df["TDS"])
```
